```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
```

```python
img = image.load_img('/content/drive/MyDrive/cnn/Train/Cat/39395293592_2ae5fd2026_n.jpg')
```

```python
plt.imshow(img)
cv2.imread('/content/drive/MyDrive/cnn/Train/Cat/39395293592_2ae5fd2026_n.jpg').shape
```

```python
#340 320
train = ImageDataGenerator(rescale=1/255)##
validation = ImageDataGenerator(rescale=1/255)
```

```python
train_dataset = train.flow_from_directory('/content/drive/MyDrive/cnn/Train/',target_size=(200,200),batch_size=3,class_mode ='binary' ) ## Train DataSet을 만들 라이브러리 설정( 라벨링 폴더에 이미지를 넣어놓아야 분류를 할수있다. )
validation_dataset = train.flow_from_directory('/content/drive/MyDrive/cnn/Test/',target_size=(200,200),batch_size=3,class_mode ='binary' ) ## Validation_DataSet 라이브러리 설정
```

```python
train_dataset.class_indices ## 라벨링한 이름과, 프로그램에서 판단할 상수
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16,(3,3),activation='relu',input_shape=(200,200,3)),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(32,(3,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    #
                                    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    #
                                    tf.keras.layers.Flatten(),
                                    ##
                                    tf.keras.layers.Dense(512,activation='relu'),
                                    tf.keras.layers.Dense(1,activation='sigmoid')
])
## 데이터 셋 모델 계층 구조 설정
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(learning_rate=0.001),
              metrics=['accuracy'])
## 설정한 설정을 모델에 적용
```

```python
model.fit = model.fit(train_dataset,
                      steps_per_epoch=5,## 한번 트레이닝 시킬때 몇번을 트레이닝 할것인지
                      epochs=30, ## 총 몇번 트레이닝 시키것인지
                      validation_data= validation_dataset
                      )
## 위에서 설정한 옵션을 본인이 구한 이미지 데이터 셋을 모델에 트레이닝 시킴
```

```python
dir_path = '/content/drive/MyDrive/cnn/testImage2/'

for i in os.listdir(dir_path):
    imgg = image.load_img(dir_path+'/'+i)
    plt.imshow(imgg)
    plt.show()

    X = image.img_to_array(imgg)
    X = np.expand_dims(X,axis=0)
    images = np.vstack([X])
    val = model.predict(X)
    ## 생성한 모델에 이미지들을 적용(디텍팅을 요청)
    if val == 0:
        print("You are Cat")
    elif val == 1:
        print("You are Dog")
    elif val == 2:
        print("You are Tiger")
    else:
        print("Who Are You??")
```
